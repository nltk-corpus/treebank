```python
































!pip install nltk
import nltk
nltk.download("all")

from nltk.tokenize import word_tokenize, sent_tokenize
sent = "Priyu is cute. she loves cats"
print(word_tokenize(sent))
print(sent_tokenize(sent))

from nltk.stem import PorterStemmer
porter = PorterStemmer()
print(porter.stem("playing"))

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("playing","v"))

from nltk import pos_tag
from nltk import word_tokenize
text = "Priyu is studying Computer Science"
tokenized_text = word_tokenize(text)
tags = tokens_tag = pos_tag(tokenized_text)
print(tags)

from nltk import word_tokenize, pos_tag, ne_chunk
nltk.download('words')
text = "Priyu is studying Computer Science"
tokenized_text = word_tokenize(text)
tags = pos_tag(tokenized_text)
entities = ne_chunk(tags)
print(entities)

from nltk import ngrams
text = "Natural Language Processing is a subfield of artificial intelligence"
n = 2
bigrams = ngrams(text.split(), n)
for grams in bigrams:
    print(grams)

from sklearn.feature_extraction.text import CountVectorizer
document = ["GeeksforGeeks is a great learning platform."]
vectorizer = CountVectorizer()
vectorizer.fit(document)
print(vectorizer.vocabulary_)
vector = vectorizer.transform(document)
print(vector.toarray())

from sklearn.feature_extraction.text import TfidfVectorizer
document = ["GeeksforGeeks is a great learning platform.",
            "It is one of the best for Computer Science Students."]
tfidf_v = TfidfVectorizer()
tfidf_matrix = tfidf_v.fit_transform(document)
print(tfidf_v.vocabulary_)
print(tfidf_matrix.toarray())
```
